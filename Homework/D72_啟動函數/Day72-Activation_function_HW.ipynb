```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return x * (x > 0)

def dReLU(x):
    return 1 * (x > 0)

x = plt.linspace(-10,10,100)
f1, = plt.plot(x, ReLU(x), "r")
f2, = plt.plot(x, dReLU(x), "b")
plt.grid()
plt.title("ReLU Function")
plt.legend([f1,f2], ["ReLU","Derivative of ReLU"], loc="upper left")
plt.show()
```
